```python
import pandas as pd
```

```python
jul_2022 = pd.read_csv('data/2022/jul2022.csv')
aug_2022 = pd.read_csv('data/2022/aug2022.csv')
sep_2022 = pd.read_csv('data/2022/sep2022.csv')
oct_2022 = pd.read_csv('data/2022/oct2022.csv')
nov_2022 = pd.read_csv('data/2022/nov2022.csv')
dec_2022 = pd.read_csv('data/2022/dec2022.csv')
```

```python
combined_2022 = pd.concat([jul_2022, aug_2022, sep_2022, oct_2022, nov_2022, dec_2022], ignore_index=True)
```

```python
combined_2022
```

```python
combined_2022 = combined_2022.drop(columns=['ride_id'])

combined_2022.head()
```

```python
combined_2022 = combined_2022.rename(columns={'started_at': 'start_time', 'ended_at': 'end_time'})

combined_2022.head()
```

```python
combined_2022.to_csv('data/2022/combined_2022.csv', index=False)
```
