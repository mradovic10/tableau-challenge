```python
import pandas as pd
```

```python
jul_2019 = pd.read_csv('data/2019/jul2019.csv')
aug_2019 = pd.read_csv('data/2019/aug2019.csv')
sep_2019 = pd.read_csv('data/2019/sep2019.csv')
oct_2019 = pd.read_csv('data/2019/oct2019.csv')
nov_2019 = pd.read_csv('data/2019/nov2019.csv')
dec_2019 = pd.read_csv('data/2019/dec2019.csv')
```

```python
combined_2019 = pd.concat([jul_2019, aug_2019, sep_2019, oct_2019, nov_2019, dec_2019], ignore_index=True)
```

```python
combined_2019
```

```python
combined_2019 = combined_2019.drop(columns=['gender', 'bikeid'])

combined_2019.head()
```

```python
combined_2019 = combined_2019.rename(columns={'starttime': 'start_time', 'stoptime': 'end_time', 'start station name': 'start_station_name', 'start station id': 'start_station_id', 'end station id': 'end_station_id', 'end station name': 'end_station_name', 'start station latitude': 'start_lat', 'start station longitude': 'start_lng', 'end station latitude': 'end_lat', 'end station longitude': 'end_lng', 'usertype': 'member_casual', 'tripduration': 'duration', 'birth year': 'birth_year'})

combined_2019.head()
```

```python
combined_2019['member_casual'] = combined_2019['member_casual'].str.replace('Subscriber', 'member')
combined_2019['member_casual'] = combined_2019['member_casual'].str.replace('Customer', 'casual')

combined_2019.head()
```

```python
combined_2019.to_csv('data/2019/combined_2019.csv', index=False)
```
